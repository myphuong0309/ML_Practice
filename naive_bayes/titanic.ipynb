```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
features = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
dummies = pd.get_dummies(features.Sex)
dummies.head(3)
```

```python
features = pd.concat([features,dummies],axis='columns')
features.head(3)
```

```python
features.drop(['Sex','male'],axis='columns',inplace=True)
features.head(3)
```

```python
features.columns[features.isna().any()]
```

```python
features.Age = features.Age.fillna(features.Age.mean())
features.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
